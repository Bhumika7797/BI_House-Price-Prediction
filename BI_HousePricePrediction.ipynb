# Data Normalization

```python
df = df.iloc[:,1:]
df_norm = (df - df.mean()) / df.std()
df_norm.head()
```

# Select X Labels

```python
X = df_norm.iloc[:, :5]
X.head()
```

#Select Y Labels

```python
Y = df_norm.iloc[:, -1]
Y.head()
```

#Feature and Label Values

```python
X_arr = X.values
Y_arr = Y.values
```

```python
print(X_arr)
print(Y_arr)
```

#Train and Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_arr, Y_arr, test_size = 0.01, shuffle = True, random_state=1)

print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

#Create the Model

```python
def get_model():

    model = Sequential([
        Dense(10, input_shape = (5,), activation = 'relu'),
        Dense(20, activation = 'relu'),
        Dense(5, activation = 'relu'),
        Dense(1)
    ])

    model.compile(
        loss='mse',
        optimizer='adadelta'
    )

    return model
```

```python
model = get_model()
model.summary()
```

```python
model = get_model()

# this prediction is before training the model
preds_on_untrained = model.predict(X_test)
```

# Model Training

```python
# Train model and store in the object history
history = model.fit(
    X_train, y_train,
    validation_data = (X_test, y_test),
    epochs = 1000
)
```

#Plot Training and Validation Loss

```python
def plot_loss(history):
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Validation'], loc='upper left')
  plt.show()
```

```python
# plot history
plot_loss(history)
```

```python
# make predictions on the trained model
preds_on_trained = model.predict(X_test)
```
